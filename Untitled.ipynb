## Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Dataset (Turbo AI).csv")
dataset.head(3)
```

## Data Pre-processing

```python
dataset.isnull().sum()
```

```python
dataset["Purchased"].unique()
```

#### Note: As we can see our data showing the classification nature, So we will use Classification Algorithms (like Logistic Regression, K-neighbors Algorithm etc) for finding the Accuracy.

```python
sns.scatterplot(data=dataset,x="Age",y="EstimatedSalary",hue="Purchased")
plt.show()
```

## Now, Dividing the dataset into Independent & Dependent variables

```python
i = dataset.iloc[:,2:-1]
d = dataset["Purchased"]
```

## Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(i)
```

```python
i = pd.DataFrame(ss.transform(i),columns=i.columns)
```

## Train-Test-Split

```python
from sklearn.model_selection import train_test_split
```

```python
i_train,i_test,d_train,d_test = train_test_split(i,d,test_size=0.2,random_state=42)
```

## Applying Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,f1_score,precision_score,recall_score
```

```python
lr = LogisticRegression()
lr.fit(i_train,d_train)
```

### Model Accuracy

```python
print("Test Accuracy :",lr.score(i_test,d_test)*100)
```

```python
print("Train Accuracy :",lr.score(i_train,d_train)*100)
```

## Confusion matrix

```python
lr_cm = confusion_matrix(d_test,lr.predict(i_test))
sns.heatmap(lr_cm,annot=True)
plt.title("Confusion matrix",fontdict={"fontsize":20,"color":"blue","fontweight":"bold"})
plt.show()
```

## F1 Score

```python
f1_score(d_test,lr.predict(i_test))*100
```

## Precision Score

```python
precision_score(d_test,lr.predict(i_test))*100
```

## Recall Score

```python
recall_score(d_test,lr.predict(i_test))*100
```

# Applying K-Neighbors Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
```

```python
for x in range(2,10):
    kn = KNeighborsClassifier(n_neighbors=x)
    kn.fit(i_train,d_train)
    print(kn.score(i_test,d_test)*100,kn.score(i_train,d_train)*100,x)
```

### Note: Testing accuracy should be less than or equal to training accuracy if not then model is Overfitted

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(i_train,d_train)
```

### Model Accuracy

```python
print("Test Accuracy :",knn.score(i_test,d_test)*100)
```

```python
print("Train Accuracy :",knn.score(i_train,d_train)*100)
```

### Confusion matrix

```python
knn_cm = confusion_matrix(d_test,knn.predict(i_test))
sns.heatmap(knn_cm,annot=True)
plt.title("Confusion Matrix",fontdict={"fontsize":20,"color":"blue","fontweight":"bold"})
plt.show()
```

### F1 Score

```python
f1_score(d_test,knn.predict(i_test))*100
```

### Precision Score

```python
precision_score(d_test,knn.predict(i_test))*100
```

### Recall Score

```python
recall_score(d_test,knn.predict(i_test))*100
```

```python
plot_decision_regions(i.to_numpy(),d.to_numpy(),clf=knn)
plt.title("K-Nearest Neighbors",fontdict={"fontsize":20,"color":"blue","fontweight":"bold"})
plt.show()
```

### As we can see K-Nearest Neighbors Algorithm giving the best accuracy for the dataset

### Final Accuracy

```python
print("Testing Accuracy :",knn.score(i_test,d_test)*100)
```

```python
print("Training Accuracy :",knn.score(i_train,d_train)*100)
```
